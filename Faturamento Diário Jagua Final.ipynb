```python
import pandas as pd

df_vendas = pd.read_excel('DATABASES\Vendas2.xlsx')
df_estados = pd.read_excel('DATABASES\Estados.xlsx')
df_filiais = pd.read_excel('DATABASES\Filiais.xlsx')
df_grupos = pd.read_excel('DATABASES\Grupos.xlsx')
print('Importações realizadas com sucesso!')

#Regras de ajustes do dataFrame
df_vendas.loc[(df_vendas['EMP_FAT']==17) & (df_vendas['GRUPO'] == '51 -SUINOS                         '), 'EMP_FAT'] = 16
df_vendas.loc[df_vendas['EMP_FAT']==2, 'DESCONTO'] = 0
df_vendas.loc[(df_vendas['EMP_FAT']==2) & (df_vendas['UF'] != 'PR') & (df_vendas['GRUPO'] == '51 -SUINOS                         '), 'EMP_FAT'] = 16
df_vendas.loc[df_vendas['EMP_FAT']==8, 'DESCONTO'] = 0

#Separando os dataFrames por dia e por mês
data_atual = df_vendas['EMISSAO'].max()
mes_atual = data_atual.month
inicio_mes = ('2021-{:02}-01'.format(mes_atual))
inicio_proximo_mes = ('2021-{:02}-01'.format(mes_atual+1))

df_vendas_dia = {}
df_vendas_mes = {}

df_vendas_dia = df_vendas.loc[df_vendas['EMISSAO'] == data_atual].copy()
df_vendas_dia['VAL_SALDO'] = df_vendas_dia['VALOR_CIF']-df_vendas_dia['VAL_DEV']-df_vendas_dia['DESCONTO']-df_vendas_dia['FRETE']
df_vendas_mes = (df_vendas.loc[(df_vendas['EMISSAO']>=inicio_mes) & (df_vendas['EMISSAO']<inicio_proximo_mes),:]).copy()
print('DataFrames criados!')
```

```python
import locale

locale.setlocale(locale.LC_NUMERIC, 'pt_br.utf-8')
pd.set_option('display.float_format', lambda x: locale.format_string('%.2f', x, grouping=True))

#Bruto
df_fat_bruto = df_vendas_dia['VALOR_CIF'].sum()
print('Faturamento Bruto: {:,.2f}'.format(df_fat_bruto))

# Devoluções
df_devolucao = df_vendas_dia['VAL_DEV'].sum()
print('Devoluções: {:,.2f}'.format(df_devolucao))
      
# Descontos
df_desconto = df_vendas_dia['DESCONTO'].sum()
print('Descontos: {:,.2f}'.format(df_desconto))
      
# Frete
df_frete = df_vendas_dia['FRETE'].sum()
print('Fretes: {:,.2f}'.format(df_frete))
      
# Líquido
df_liquido = df_fat_bruto - df_devolucao - df_desconto - df_frete
print('Faturamento Líquido: {:,.2f}'.format(df_liquido))


#Vendas por filial
df_vendas_filiais = df_vendas_dia[['EMP_FAT','QTD_SALDO','VAL_SALDO']]
df_vendas_filiais = df_vendas_filiais.merge(df_filiais, on='EMP_FAT')
filiais = df_vendas_filiais.groupby('FILIAL').sum()
filiais = filiais.drop('EMP_FAT', 1)
filiais['P.M.'] = filiais['VAL_SALDO'] / filiais['QTD_SALDO']
qtd_total = filiais['QTD_SALDO'].sum()
saldo_total = filiais['VAL_SALDO'].sum()
pm_total = saldo_total / qtd_total
filiais.loc['TOTAL'] = (qtd_total,saldo_total,pm_total)
display(filiais)
filiais.to_html('DataFrames\Vendas Por Filial.html')


#Vendas por segmento
df_vendas_segmento = df_vendas_dia[['EMP_FAT','ITEM','QTD_SALDO','VAL_SALDO']]
df_vendas_segmento = df_vendas_segmento.merge(df_grupos, on='ITEM')
segmento = df_vendas_segmento.groupby(['TIPO']).sum()
segmento['P.M.'] = segmento['VAL_SALDO'] / segmento['QTD_SALDO']
segmento = segmento.drop('EMP_FAT', 1)
qtd_total = segmento['QTD_SALDO'].sum()
saldo_total = segmento['VAL_SALDO'].sum()
pm_total = saldo_total / qtd_total
segmento.loc['TOTAL'] = (qtd_total,saldo_total,pm_total)
display(segmento)
segmento.to_html('DataFrames\Vendas Por Segmento.html')

#Vendas por segmento e mercado
df_vendas_segmento_mercado = df_vendas_dia[['EMP_FAT','ITEM','QTD_SALDO','VAL_SALDO']]
df_vendas_segmento_mercado = df_vendas_segmento_mercado.merge(df_grupos, on='ITEM')
df_vendas_segmento_mercado['TIPO_MERCADO'] = df_vendas_segmento_mercado['TIPO'] + ' - ' + df_vendas_segmento_mercado['MERCADO']
segmento_mercado = df_vendas_segmento_mercado.groupby(['TIPO_MERCADO']).sum()
segmento_mercado['P.M.'] = segmento_mercado['VAL_SALDO'] / segmento_mercado['QTD_SALDO']
segmento_mercado = segmento_mercado.drop('EMP_FAT', 1)
qtd_total = segmento_mercado['QTD_SALDO'].sum()
saldo_total = segmento_mercado['VAL_SALDO'].sum()
pm_total = saldo_total / qtd_total
segmento_mercado.loc['TOTAL'] = (qtd_total,saldo_total,pm_total)
display(segmento_mercado)
segmento_mercado.to_html('DataFrames\Vendas Por Segmento e Mercado.html')

#Vendas por segmento e mercado
df_vendas_outras = df_vendas_dia[['EMP_FAT','ITEM','QTD_SALDO','VAL_SALDO']]
df_vendas_outras = df_vendas_outras.merge(df_grupos, on='ITEM')
df_vendas_outras = df_vendas_outras.loc[df_vendas_outras['TIPO'] == 'OUTRAS']
df_vendas_outras['OUTRAS_VENDAS'] = df_vendas_outras['TIPO'] + ' - ' + df_vendas_outras['GRUPO']
segmento_outras = df_vendas_outras.groupby(['OUTRAS_VENDAS']).sum()
segmento_outras['P.M.'] = segmento_outras['VAL_SALDO'] / segmento_outras['QTD_SALDO']
segmento_outras = segmento_outras.drop('EMP_FAT', 1)
qtd_total = segmento_outras['QTD_SALDO'].sum()
saldo_total = segmento_outras['VAL_SALDO'].sum()
pm_total = saldo_total / qtd_total
segmento_outras.loc['TOTAL'] = (qtd_total, saldo_total, pm_total)
display(segmento_outras)
segmento_outras.to_html('DataFrames\Outras Vendas.html')


#Vendas por Filial e Local de Produção (Lado esquerdo do relatório)

empresas = ['JAGUAPITÃ', 'RONDON', 'NOVA PRATA', 'IPUAÇU', 'CURITIBA', 'ARAPONGAS', 'SÃO PAULO', 'XANXERE', 'RIO G. SUL']
i = 0

for empresa in empresas:
        print('\n')
        print(empresas[i])
        df_empresa = df_vendas_dia[['EMP_FAT','ITEM','QTD_SALDO','VAL_SALDO']]
        df_empresa = df_empresa.merge(df_grupos, on='ITEM')
        df_empresa = df_empresa.merge(df_filiais, on='EMP_FAT')
        df_empresa = df_empresa.loc[df_empresa['FILIAL'] == empresa]
        df_empresa = df_empresa.loc[(df_empresa['TIPO'] != 'SUBPRODUTOS AVES') & \
                                    (df_empresa['TIPO'] != 'SUBPRODUTOS SUÍNOS') & \
                                    (df_empresa['TIPO'] != 'OUTRAS')]
        df_empresa['FILIAL_MERCADO_TIPO'] = df_empresa['FILIAL'] + ' - ' + df_empresa['TIPO'] + ' - ' + df_empresa['MERCADO']
        df_empresa = df_empresa.groupby(['FILIAL_MERCADO_TIPO','LOCAL_PRODUCAO']).sum()
        df_empresa = df_empresa.drop('EMP_FAT', 1)
        df_empresa['P.M.'] = df_empresa['VAL_SALDO'] / df_empresa['QTD_SALDO']
        i += 1
        display(df_empresa)
        print('\n')
        df_empresa.to_html('DataFrames\Empresa - {}.html'.format(empresas[i-1]))

#Vendas por região
df_vendas_regiao = df_vendas_dia[['EMP_FAT','UF','VAL_SALDO']]
df_vendas_regiao = df_vendas_regiao.merge(df_estados, on='UF')
df_vendas_regiao = df_vendas_regiao.drop('EMP_FAT', 1)
df_vendas_regiao = df_vendas_regiao.drop('UF', 1)
df_vendas_regiao = df_vendas_regiao.drop('ESTADO', 1)
df_vendas_regiao = df_vendas_regiao.groupby('REGIÃO').sum()
df_vendas_regiao = df_vendas_regiao.sort_values(by='VAL_SALDO', ascending=False)

display(df_vendas_regiao)
df_vendas_regiao.plot.pie(subplots=True, legend=False, figsize=(5,5),  ylabel="", autopct='%1.f%%')        
```
