```python
import pandas as pd

df_vendas = pd.read_excel('DATABASES\Vendas.xlsx')
df_estados = pd.read_excel('DATABASES\Estados.xlsx')
df_filiais = pd.read_excel('DATABASES\Filiais.xlsx')
df_grupos = pd.read_excel('DATABASES\Grupos.xlsx')
print('Importações realizadas com sucesso!')

data_atual = df_vendas['EMISSAO'].max()
df_vendas_dia = {}

df_vendas_dia = df_vendas.loc[df_vendas['EMISSAO'] == data_atual].copy()
df_vendas_dia['VAL_SALDO'] = df_vendas_dia['VALOR_CIF']-df_vendas_dia['VAL_DEV']-df_vendas_dia['DESCONTO']-df_vendas_dia['FRETE']
print('DataFrames criados!')
print(data_atual)
```

## Sem Subtotais - GroupBy

```python
import locale

cell_hover = {  # for row hover use <tr> instead of <td>
    'selector': 'td:hover',
    'props': [('background-color', '#ffffb3')]
}
index_names = {
    'selector': '.index_name',
    'props': 'font-style: italic; color: darkgrey; font-weight:normal;'
}
headers = {
    'selector': 'th:not(.index_name)',
    'props': 'background-color: #000066; color: white;'
}

locale.setlocale(locale.LC_NUMERIC, 'pt_br.utf-8')
pd.set_option('display.float_format', lambda x: locale.format_string('%.2f', x, grouping=True))

emp_fat = df_vendas_dia['EMP_FAT'].unique()
emp_fat.sort()
print(emp_fat)

for emp in emp_fat:
    df_01 = df_vendas_dia.loc[df_vendas_dia['EMP_FAT'] == emp].copy()
    df_01 = df_01[['FINALID', 'FINALIDADE', 'ITEM', 'DESC_ITEM','QTD_CX', 'PESO','PRECO_CIF', 'VALOR_CIF', 'DESCONTO', 'FRT_UNIT', 'FRETE','FRETE VDJ', 'VALOR_FOB', 'PRECO_FOB','QTD_DEV', 'VAL_DEV', 'QTD_SALDO', 'VAL_SALDO', 'ICMS TRIB', 'PIS TRIB','COFINS TRIB']]
    df_01 = df_01.groupby(['FINALID', 'FINALIDADE', 'ITEM', 'DESC_ITEM']).\
        agg(QTD_CAIXA=('QTD_CX','sum'),PESO=('PESO','sum'),\
        PRECO_MEDIO=('PRECO_CIF','mean'),VALOR_CIF=('VALOR_CIF','sum'),\
        DESCONTO=('DESCONTO','sum'),FRETE=('FRETE','sum'),\
        VALOR_FOB=('VALOR_FOB','sum'),PRECO_FOB=('PRECO_FOB','mean'),\
        PRECO_MAX=('PRECO_FOB','max'),PRECO_MIN=('PRECO_FOB','min'),\
        QTD_DEV=('QTD_DEV','sum'),VAL_DEV=('VAL_DEV','sum'),\
        QTD_SALDO=('QTD_SALDO','sum'),VAL_SALDO=('VAL_SALDO','sum'),\
        ICMS_TRIB=('ICMS TRIB','sum'),PIS_TRIB=('PIS TRIB','sum'),COFINS_TRIB=('COFINS TRIB','sum'))
    display(df_01)
    df_01.to_html('Apoio\Faturamento Diário\Empresa {} - {}-{}-{}.html'.format(emp, data_atual.day, data_atual.month, data_atual.year))
print('Exportações concluídas!')
```

## Apenas Total Geral - Pivot_table

```python
import numpy as np
import locale

locale.setlocale(locale.LC_NUMERIC, 'pt_br.utf-8')
pd.set_option('display.float_format', lambda x: locale.format_string('%.2f', x, grouping=True))

emp_fat = df_vendas_dia['EMP_FAT'].unique()
emp_fat.sort()
print(emp_fat)

for emp in emp_fat:
    df_faturamento = df_vendas_dia.loc[df_vendas_dia['EMP_FAT'] == emp].copy()
    df_faturamento = pd.pivot_table(df_faturamento, 
                                    index=['FINALID', 'FINALIDADE', 'ITEM', 'DESC_ITEM'],
                                    values=['QTD_CX', 'PESO',
                                           'PRECO_CIF', 'VALOR_CIF', 'DESCONTO', 'FRETE',
                                           'VALOR_FOB', 'PRECO_FOB',
                                           'QTD_DEV', 'VAL_DEV', 'QTD_SALDO', 'VAL_SALDO', 'ICMS TRIB', 'PIS TRIB',
                                           'COFINS TRIB'],
                                    aggfunc={'QTD_CX':np.sum, 'PESO':np.sum, 'PRECO_CIF':np.mean,
                                    'VALOR_CIF':np.sum,'DESCONTO':np.sum,'FRETE':np.sum,
                                    'VALOR_FOB':np.sum,'PRECO_FOB':np.mean, 'PRECO_FOB':np.max,'PRECO_FOB':np.min,
                                    'QTD_DEV':np.sum,'VAL_DEV':np.sum, 'QTD_SALDO':np.sum, 'VAL_SALDO':np.sum,
                                    'ICMS TRIB':np.sum,'PIS TRIB':np.sum, 
                                    'COFINS TRIB':np.sum}, margins=True, margins_name='Total', dropna=True)
    print(emp)
    display(df_faturamento)

df_faturamento.to_excel('Faturamento por empresa.xlsx')
```

## Subtotais com NaN - Em desenvolvimento

```python
df1 = df_vendas_dia.copy()

pt = pd.pivot_table(df_vendas_dia,
                    index=['FINALID', 'FINALIDADE'],
                    values=['QTD_CX', 'PESO',
                            'PRECO_CIF', 'VALOR_CIF', 'DESCONTO', 'FRETE',
                            'VALOR_FOB', 'PRECO_FOB','QTD_DEV', 'VAL_DEV', 'QTD_SALDO',
                            'VAL_SALDO', 'ICMS TRIB', 'PIS TRIB','COFINS TRIB'],
                    aggfunc={'QTD_CX':np.sum, 'PESO':np.sum, 'PRECO_CIF':np.mean,
                             'VALOR_CIF':np.sum,'DESCONTO':np.sum,'FRETE':np.sum,
                             'VALOR_FOB':np.sum,'PRECO_FOB':np.mean, 'PRECO_FOB':np.max,'PRECO_FOB':np.min,
                             'QTD_DEV':np.sum,'VAL_DEV':np.sum, 'QTD_SALDO':np.sum, 'VAL_SALDO':np.sum,
                             'ICMS TRIB':np.sum,'PIS TRIB':np.sum, 'COFINS TRIB':np.sum}, 
                    margins=True,
                    margins_name='TOTAL GERAL',
                    fill_value=0,
                    dropna=True)

df1 = df1.groupby(['FINALID']).agg(QTD_CX=('QTD_CX','sum'),PESO=('PESO','sum'),
                                      PRECO_CIF=('PRECO_CIF','mean'),VALOR_CIF=('VALOR_CIF','sum'),
                                      DESCONTO=('DESCONTO','sum'),FRETE=('FRETE','sum'),
                                      VALOR_FOB=('VALOR_FOB','sum'),PRECO_FOB=('PRECO_FOB','mean'),
                                      PRECO_MAX=('PRECO_FOB','max'),PRECO_MIN=('PRECO_FOB','min'),
                                      QTD_DEV=('QTD_DEV','sum'),VAL_DEV=('VAL_DEV','sum'),
                                      QTD_SALDO=('QTD_SALDO','sum'),VAL_SALDO=('VAL_SALDO','sum'),
                                      ICMS_TRIB=('ICMS TRIB','sum'),PIS_TRIB=('PIS TRIB','sum'),
                                      COFINS_TRIB=('COFINS TRIB','sum'))


df1.index = pd.MultiIndex.from_arrays([df1.index + ' - SUBTOTAL', len(df1.index) * ['']])


display(pd.concat([df1, pt]).sort_index())
pd.concat([pt, df1]).sort_index().to_html('Fat por empresa.html')
```
